# Let's Analyze all the companies publish rate and view count with likes in it

# Time Series Plot

**For Zee Music Company**

```python
plt.figure(figsize=(12, 8))  
df['published_at'] = pd.to_datetime(df['published_at'])
df = df.sort_values('published_at')  
videos_published_df = df['published_at'].dt.to_period('M').value_counts().sort_index()
plt.plot(videos_published_df.index.to_timestamp(), videos_published_df.values, marker='o')
plt.title('Videos Published Over Time (Zee Music Company)')
plt.xlabel('Published Date')
plt.ylabel('Number of Videos Published')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Let's compare for all three companies (i.e Zee Company , T-Series and Tips Official)

# KDE Plot For Distribution of View count

```python
df['view_count'] = pd.to_numeric(df['view_count'], errors='coerce')
df1['view_count'] = pd.to_numeric(df1['view_count'], errors='coerce')
df2['view_count'] = pd.to_numeric(df2['view_count'], errors='coerce')
df = df.dropna(subset=['view_count'])
df1 = df1.dropna(subset=['view_count'])
df2 = df2.dropna(subset=['view_count'])
plt.figure(figsize=(10, 6))
sns.kdeplot(df['view_count'], shade=True, label='Zee Music Company', color='blue')
sns.kdeplot(df1['view_count'], shade=True, label='T-Series', color='red')
sns.kdeplot(df2['view_count'], shade=True, label='Tips Official', color='green')
plt.title('Distribution of View Counts')
plt.xlabel('View Count')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Scatterplot


# Top 20 highest commented video

```python
combined_df = pd.concat([df, df1, df2])
top_20_commented = combined_df.nlargest(20, 'comment_count')
def extract_first_two_words(title):
    words = title.split()
    if len(words) >= 5:
        return ' '.join(words[:5])
    else:
        return title
top_20_commented['short_title'] = top_20_commented['title'].apply(extract_first_two_words)
plt.figure(figsize=(12, 8))
plt.scatter(top_20_commented['view_count'], top_20_commented['comment_count'], color='purple', alpha=0.7)
plt.title('Top 20 Highest Commented Videos')
plt.xlabel('View Count')
plt.ylabel('Comment Count')
plt.grid(True)
for i, row in top_20_commented.iterrows():
    plt.text(row['view_count'], row['comment_count'], row['short_title'], fontsize=8, ha='right', va='bottom')
plt.tight_layout()
plt.show()
```

# Top 10 Highest Viewed Video

```python
top_10_viewed = combined_df.nlargest(10, 'view_count')
def extract_first_two_words(title):
    words = title.split()
    if len(words) >= 5:
        return ' '.join(words[:5])
    else:
        return title
top_10_viewed['short_title'] = top_10_viewed['title'].apply(extract_first_two_words)
plt.figure(figsize=(12, 8))
plt.scatter(top_10_viewed['view_count'], top_10_viewed['comment_count'], color='blue', alpha=0.7)
plt.title('Top 10 Highest Viewed Videos')
plt.xlabel('View Count')
plt.ylabel('Comment Count')
plt.grid(True)
for i, row in top_10_viewed.iterrows():
    plt.text(row['view_count'], row['comment_count'], row['short_title'], fontsize=8, ha='right', va='bottom')
plt.tight_layout()
plt.show()
```

# Top 10 Highest Liked Video

```python
top_10_liked = combined_df.nlargest(10, 'like_count')
def extract_first_two_words(title):
    words = title.split()
    if len(words) >= 5:
        return ' '.join(words[:5])
    else:
        return title
top_10_liked['short_title'] = top_10_liked['title'].apply(extract_first_two_words)
plt.figure(figsize=(12, 8))
plt.scatter(top_10_liked['like_count'], top_10_liked['comment_count'], color='green', alpha=0.7)
plt.title('Top 10 Highest Liked Videos')
plt.xlabel('Like Count')
plt.ylabel('Comment Count')
plt.grid(True)
for i, row in top_10_liked.iterrows():
    plt.text(row['like_count'], row['comment_count'], row['short_title'], fontsize=8, ha='right', va='bottom')
plt.tight_layout()
plt.show()
```
